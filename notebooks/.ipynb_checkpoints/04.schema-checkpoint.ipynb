```python
from pandas import DataFrame, read_parquet

# Load the cleaned dataset
df: DataFrame = read_parquet("../data/curated/RWventas.parquet", engine="pyarrow")

#
df_columns = df.columns.tolist()
df_columns.remove("tipo_producto")
sales_columns = [
    "city",
    "date",
    "product",
    "quantity",
    "unit_price",
    "type_sales",
    "type_clients",
    "discount",
    "shipping_cost",
    "total_price",
]

# sales
types_sales = (
    DataFrame(df["tipo_venta"].unique(), columns=["type_sales"])
    .reset_index()
    .rename(columns={"index": "id_types_sales"})
)

# clients
types_clients = (
    DataFrame(df["tipo_cliente"].unique(), columns=["type_clients"])
    .reset_index()
    .rename(columns={"index": "id_types_clients"})
)

# products
types_products = (
    DataFrame(df["tipo_producto"].dropna().unique(), columns=["type_product"])
    .reset_index()
    .rename(columns={"index": "id_type_product"})
)

products = (
    df[["producto", "tipo_producto"]]
    .drop_duplicates()
    .rename(columns={"producto": "product", "tipo_producto": "type_product"})
    .merge(types_products, on="type_product", how="left")
    .reset_index()
    .rename(columns={"index": "id_product"})[
        ["id_product", "product", "id_type_product"]
    ]
)

df_rename = df.rename(columns=dict(zip(df_columns, sales_columns)))
sales = (
    df_rename.merge(types_sales, on="type_sales", how="left")
    .merge(types_clients, on="type_clients", how="left")
    .merge(products, on="product", how="left")
    .loc[:, sales_columns]
)

# Save the dimension tables and fact table

```
